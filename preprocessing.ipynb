```python
import pandas as pd
import numpy as np
import re
import unicodedata
from underthesea import text_normalize, word_tokenize
```

```python
# Đọc dữ liệu
df = pd.read_csv('data/data.csv')
df
```

```python
# Xoá html code trong dữ liệu
def remove_html(text):
    return re.sub(r'<[^>]*>', '', text)
```

```python
# Chuẩn hoá về bảng mã Unicode, kiểu gõ dấu tiếng việt và chữ viết thường
def normalize(text):
    text = unicodedata.normalize('NFC', text)
    text = text_normalize(text)
    text = text.lower()
    return text
```

```python
# Xoá các ký tự đặc biệt và các khoảng cách thừa
def remove_special_characters(text):
    text = re.sub(r'[^\w\sàáảãạăắằẳẵặâấầẩẫậđèéẻẽẹêếềểễệìíỉĩịòóỏõọôốồổỗộơớờởỡợùúủũụưứừửữựỳýỷỹỵ]', ' ', text)
    text = re.sub(r'\s+', ' ', text)
    return text.strip()
```

```python
# Loại bỏ stopword
with open('data/vietnamese-stopwords.txt', encoding='utf-8') as f:
    stopwords = set(f.read().splitlines())
    
def remove_stopwords(text):
    tokens = word_tokenize(text)
    text_filterd = [word for word in tokens if word not in stopwords]
    return ' '.join(text_filterd)
```

```python
def preprocessing(texts):
    result = []
    for text in texts:
        text = remove_html(text)
        text = normalize(text)
        text = remove_special_characters(text)
        text = remove_stopwords(text)
        result.append(text)
    return result
```

```python
result = preprocessing(df['texts'])
result
```

```python
df['clean_texts'] = result
df
```
